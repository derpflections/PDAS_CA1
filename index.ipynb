## Data Cleaning/Wrangling
Defining array of variables for use later

```python
catSec, cat1num, cat2num, cat3num, cat4num, cat5num, cat6num, cat7num = [], [], [], [], [], [], [], []
arr2008, arr2009, arr2010, arr2011, arr2012, arr2013, arr2014, arr2015, arr2016, arr2017, arr2018, arr2019, arr2020 = [], [], [], [], [], [], [], [], [], [], [], [], []
polyFemale, polyMale, preuMale, preuFemale, iteMale, iteFemale = [], [], [], [], [], []
cat1Male, cat1Female, cat2Male, cat2Female, cat3Male, cat3Female, cat4Male, cat4Female, cat5Male, cat5Female, cat6Male, cat6Female, cat7Male, cat7Female = [], [], [], [], [], [], [], [], [], [], [], [], [], []
secondaryMale, secondaryFemale = [], []
pct1, pct2, pct3, pct4, pct5, pct6, pct7 = [] ,[], [], [], [], [], []
catAllgradPct, cat1gradPct, cat2gradPct, cat3gradPct, cat4gradPct, cat5gradPct, cat6gradPct, cat7gradPct = [], [], [], [], [], [], [], []
catNumber = [cat1num, cat2num, cat3num, cat4num, cat5num, cat6num, cat7num, catSec]
yearNumber = [arr2009, arr2010, arr2011, arr2012, arr2013, arr2014, arr2015, arr2016, arr2017, arr2018, arr2019, arr2020]
year = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
categoryNames = ['Science', 'Arts', 'Business', 'Infocomm', 'Engineering', 'Service', 'Social']
genderArray = [cat1Male, cat1Female, cat2Male, cat2Female, cat3Male, cat3Female, cat4Male, cat4Female, cat5Male, cat5Female, cat6Male, cat6Female, cat7Male, cat7Female]
typeGenderArray = [polyFemale, polyMale, preuMale, preuFemale, iteMale, iteFemale]
maleArray = [cat1Male, cat2Male, cat3Male, cat4Male, cat5Male, cat6Male, cat7Male]
femaleArray = [cat1Female, cat2Female, cat3Female, cat4Female, cat5Female, cat6Female, cat7Female]
genderPct = [pct1, pct2, pct3, pct4, pct5, pct6, pct7]
gradPct = [cat1gradPct, cat2gradPct, cat3gradPct, cat4gradPct, cat5gradPct, cat6gradPct, cat7gradPct]
```

Defining Secondary Category

```python
for i in range (2008, 2020):
    targetArr = secondaryClean
    filter1 = (targetArr['year'] == i) & (targetArr['level'] == 'Secondary 4') & (targetArr['gender'] == 'MF')
    catSec.append(secondaryClean[filter1]['enrolment'].sum())
print(catSec)
```

Defining filters for tertiary datasets.

```python
def catFilter(targetArr, year, useFilter, gender):
    cat1filter = (targetArr['year'] == year) & ((targetArr['course'] == 'Applied and Health Sciences') | (targetArr['course'] == 'Health Sciences') | (targetArr['course'] == 'Natural & Mathematical Sciences') | (targetArr['course'] == 'Natural & Physical & Mathematical Sciences') | (targetArr['course'] == 'Science')) & (targetArr['gender'] == gender)
    cat2filter = (targetArr['year'] == year) & ((targetArr['course'] == 'Arts') | (targetArr['course'] == 'Applied Arts') | (targetArr['course'] == 'Architecture & Bulding & Real Estate') | (targetArr['course'] == 'Design & Media')) & (targetArr['gender'] == gender)
    cat3filter = (targetArr['year'] == year) & ((targetArr['course'] == 'Business & Services') | (targetArr['course'] == 'Business & Administration') | (targetArr['course'] == 'Commerce')) & (targetArr['gender'] == gender)
    cat4filter = (targetArr['year'] == year) & ((targetArr['course'] == 'Electronic & Infocomm Technology') | (targetArr['course'] == 'Information Technology')) & (targetArr['gender'] == gender)
    cat5filter = (targetArr['year'] == year) & ((targetArr['course'] == 'Engineering') | (targetArr['course'] == 'Engineering Sciences')) & (targetArr['gender'] == gender)
    cat6filter = (targetArr['year'] == year) & ((targetArr['course'] == 'Hospitality') | (targetArr['course'] == 'Services')) & (targetArr['gender'] == gender)
    cat7filter = (targetArr['year'] == year) & ((targetArr['course'] == 'Humanities & Social Sciences') | (targetArr['course'] == 'Law') | (targetArr['course'] == 'Mass Communication')) & (targetArr['gender'] == gender)
    filterArray = [cat1filter, cat2filter, cat3filter, cat4filter, cat5filter, cat6filter, cat7filter]
    return targetArr[filterArray[useFilter]]
```

Breaking the courses down into 7 categories.
>(Science (1), Arts (2), Business (3), Infocomm (4), Engineering (5), Service (6), Social (7). )

```python
for i in range (2009, 2021):
    for categoryNo in range(0,7):
        reference = 0 
        reference += int(catFilter(iteClean, i, categoryNo, "MF")['enrolment'].sum())
        reference += int(catFilter(polyClean, i, categoryNo, "MF")['enrolment'].sum())
        reference += int(catFilter(preuClean, i, categoryNo, "MF")['enrolment'].sum())
        catNumber[categoryNo].append(reference)
```

Breaking the courses down by year, from 2008 to 2020.

```python
for e in range(0,12):
    tempArr = []
    for i in range(0,7):
        yearNumber[e].append(catNumber[i][e])
```

Breaking the courses down by gender

```python
for i in range (2008, 2019):
    secondaryGenderFilterF = (secondaryData['gender'] == "F") & (secondaryData["level"] == "Secondary 4") & (secondaryData['year'] == i)
    secondaryGenderFilterMF = (secondaryData['gender'] == "MF") & (secondaryData["level"] == "Secondary 4") & (secondaryData['year'] == i)
    tempMF = secondaryData[secondaryGenderFilterMF]["enrolment"].sum()
    tempFem = secondaryData[secondaryGenderFilterF]["enrolment"].sum()
    tempM = tempMF - tempFem
    secondaryFemale.append(tempFem)
    secondaryMale.append(tempM)

for i in range (2009, 2021):
    for categoryNo in range(0,7):
        tempMF, tempF, tempM = 0,0,0
        tempF += catFilter(polyClean, i, categoryNo, "F")["enrolment"].sum()
        tempF += catFilter(iteClean, i, categoryNo, "F")["enrolment"].sum()
        tempF += catFilter(preuClean, i, categoryNo, "F")["enrolment"].sum()
        tempMF += catFilter(polyClean, i, categoryNo, "MF")["enrolment"].sum()
        tempMF += catFilter(iteClean, i, categoryNo, "MF")["enrolment"].sum()
        tempMF += catFilter(preuClean, i, categoryNo, "MF")["enrolment"].sum()
        tempM = tempMF - tempF
        femaleArray[categoryNo].append(tempF)
        maleArray[categoryNo].append(tempM)

for i in range(0, len(year[1:])):
    for categoryNo in range(0,7):
        genderPct[categoryNo].append(((femaleArray[categoryNo][i]/catNumber[categoryNo][i])*100).round(3))
```

Computing the percentage of graduates for each course.

```python
for i in range(2009, 2021):
    for categoryNo in range(0,7): 
        gradPct[categoryNo].append(((catFilter(polyClean, i, categoryNo, "MF")["graduates"].sum()+catFilter(iteClean, i, categoryNo, "MF")["graduates"].sum())/ (catFilter(polyClean, i, categoryNo, "MF")["enrolment"].sum() +catFilter(iteClean, i, categoryNo, "MF")["enrolment"].sum())*100).round(4))

for categoryNo in range(0,7):
    catAllgradPct.append(gradPct[categoryNo])
catAllgradPctNp = np.array(catAllgradPct)
catAllgradPctNp = catAllgradPctNp.flatten()
```
